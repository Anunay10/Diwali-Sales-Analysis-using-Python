```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali_Sales.csv",encoding='unicode_escape')
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
df.head(7)
```

```python
df["Amount"]=df["Amount"].astype("int")
```

```python
df.info()
```

```python
#Exploratory Data Analysis 
```

```python
df.head()
```

```python
sns.countplot(x="Gender",data=df)
plt.title("Gender counts")
```

```python
Sales_per_Gender=df.groupby("Gender")[["Amount"]].sum().sort_values("Amount",ascending=False).reset_index()
Sales_per_Gender
```

```python
sns.barplot(x="Gender",y="Amount",data=Sales_per_Gender)
plt.title("Who is spending more M or F ?")
```

```python
#From the above graphs, we can see that females tend to shop more frequent and also spends more than males
```

```python
df.columns
```

```python
sns.countplot(x="Age Group",data=df,hue="Gender")
plt.title("Males and Females counts in each Groups")
```

```python
# This graphs shows that females in the age groups 26-35 are shopping more
```

```python
AgeGroup_Analysis=df.groupby("Age Group")[["Amount"]].sum().sort_values("Amount",ascending=False).reset_index()
AgeGroup_Analysis
```

```python
sns.barplot(x="Age Group",y="Amount",data=AgeGroup_Analysis)
plt.title("Age Groups buying Potentials")
```

```python
# from the above graphs we can see that Females in age groups 26-35 are shopping more than others
```

```python
df.head()
```

```python
Orders_by_States=df.groupby("State")[["Orders"]].sum().sort_values("Orders",ascending=False).reset_index().head(10)
Orders_by_States
```

```python
sns.barplot(x="State",y="Orders",data=Orders_by_States)
sns.set(rc={"figure.figsize": (20, 10)})
plt.title("Orders per States")
```

```python
# From the above graph, we can conclude that the people in UP have no. of orders more
```

```python
Amount_per_State=df.groupby("State")[["Amount"]].sum().sort_values("Amount",ascending=False).reset_index().head(10)
Amount_per_State
```

```python
sns.barplot(x="State",y="Amount",data=Amount_per_State)
plt.title("Total Shopping Amonnt that states are spending during Daiwali")
```

```python
# From the above graphs we can observe that people living in UP are odering more as well as spending more
#Also even though Ker is ordering more but Har is spending more than Ker
```

```python
sns.countplot(x="Marital_Status",data=df,hue="Gender")
sns.set(rc={"figure.figsize": (10, 4)})
plt.title("Married Count")
```

```python
Married_spent=df.groupby("Marital_Status")[["Amount"]].sum().sort_values("Amount",ascending=False).reset_index()
Married_spent
```

```python
sns.countplot(x="Occupation",data=df,hue="Gender")
sns.set(rc={"figure.figsize": (25, 4)})
plt.title("Gender distribution amongst Occupation")
```

```python
Occupation_per_Spending=df.groupby("Occupation")[["Amount"]].sum().sort_values("Amount",ascending=False).reset_index()
sns.barplot(x="Occupation",y="Amount",data=Occupation_per_Spending)
plt.title("Different Occupation Spending Power")
```

```python
# from the above graphs, we can conclude that females from IT sector are spending more 
```
